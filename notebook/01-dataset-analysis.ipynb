```python
from torchvision.datasets import CIFAR10
```

```python
train_dataset = CIFAR10(
    root="../data",
    train=True,
    download=True
)

test_dataset = CIFAR10(
    root="../data",
    train=False,
    download=True
)
```

```python
print(len(train_dataset))
print(len(test_dataset))
```

```python
print(train_dataset.classes)
```

```python
import numpy as np
import pandas as pd

labels = np.array(train_dataset.targets)

df = pd.DataFrame({
    "label": labels
})

df["class_name"] = df["label"].apply(
    lambda x: train_dataset.classes[x]
)

df.head()
```

```python
class_counts = df["class_name"].value_counts()

print(class_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

class_counts.plot(kind="bar")

plt.title("Class Distribution in CIFAR-10")
plt.xlabel("Class")
plt.ylabel("Number of Images")

plt.show()
```

```python
def show_images(class_index, num_images=5):

    plt.figure(figsize=(15,3))

    count = 0

    for img, label in train_dataset:

        if label == class_index:

            plt.subplot(1, num_images, count+1)
            plt.imshow(img)
            plt.axis("off")

            count += 1

            if count == num_images:
                break

    plt.show()
```

```python
show_images(0) #airplane
```

```python
show_images(3) #cats
```

```python
show_images(5) #dogs
```

```python
show_images(9) #truck
```
